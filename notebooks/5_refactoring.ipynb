# Refactoring

When writing this course, I kept returning to one question: should we teach software design from a clean slate, or from the far more common reality of inheriting messy code under time pressure?

For this lecture, we assume the second case. You already have a codebase, you are already using it for science, and there is rarely time or budget for a full rewrite. So the goal here is practical: how to improve what you have, safely and incrementally.

> *Definition*: **Refactoring** is improving the internal structure of code without changing external behaviour. It makes code easier to read, test, and extend.

I also avoid showing intentionally bad code examples here; those snippets are often copied into real projects. The goal is to build better habits, not seed new anti-patterns.

## Anti-patterns

Messy legacy code is rarely the result of lazy people. More often, it is the natural consequence of **time pressure, shifting goals, and shortcuts that once looked reasonable**.

> *Definition*: An **anti-pattern** is a common but counterproductive solution to a recurring problem.

Typical signs:

- You see it repeatedly.
- It feels convenient in the short term.
- It creates long-term pain (bugs, complexity, technical debt).
- Better alternatives exist, but require more structure upfront.

Examples include excessive global state, copy-paste logic, and overcomplicated abstractions. The key point: a pattern becomes an anti-pattern when short-term convenience hides long-term harm.

## Technical debt

Technical debt is the accumulation of design shortcuts, quick fixes, and outdated structures that make future work harder.

> *Definition*: **Technical debt** is the cost of past shortcuts, paid later through slower development, harder debugging, and riskier changes.

In academia, this debt is often invisible because no one explicitly budgets for refactoring. Yet someone always pays. One of the biggest hidden costs is onboarding: each new team member (student, postdoc, etc.) inherits architectural compromises before they can contribute new science.

A useful mental model: **onboarding lag is interest on technical debt**.

Refactoring is not busywork. It is one of the few deliberate ways to reduce that debt and improve long-term scientific productivity.

## How to refactor

Refactoring legacy code means gradually turning a fragile but working system into one that is clearer and safer&mdash;without changing what it does.

A practical process:

1. **Start with a safety net**
    - Add minimal tests for current behaviour.
    - Commit to version control.
    - Work in small, reversible steps.

2. **Target hotspots first**
    - Code you touch often.
    - Bug-prone areas.
    - Hard-to-understand modules.

3. **Make the smallest useful improvements**
    - Split oversized functions.
    - Remove duplication.
    - Delete dead code.
    - Isolate global state.

4. **Replace anti-patterns deliberately**
    - Copy-paste blocks -> shared helpers.
    - Scattered config -> structured config objects.
    - Deep nesting -> clearer control flow.

5. **Improve names and structure**
    - Rename unclear identifiers.
    - Group related code.
    - Keep functions focused.

6. **Add tests as you go**
    - Unit tests for clarified functions.
    - Regression tests for known bug paths.
    - Edge-case checks once behaviour is understood.

7. **Treat refactoring as a habit**
    - Leave touched code slightly better each time.
    - Avoid "big rewrite" fantasies unless truly necessary.

**The goal is not perfection**. It is to steadily reduce friction so future you (and future collaborators) can work with confidence.

## Refactoring vs code creep

In academia, refactoring competes directly with research time. That makes it essential to distinguish useful refactoring from open-ended "cleanup" that expands uncontrollably.

You are refactoring if:

- External behaviour stays the same.
- The structural improvement is clear.
- The task has a clear stopping point.
- Future maintenance becomes easier.

You are doing code creep if:

- Scope keeps expanding while "cleaning up."
- You add new features during restructuring.
- You cannot define when the task is done.

In short:

- Refactoring is deliberate and time-boxed.
- Code creep is open-ended and risky.

### Time-boxing rule

Set a fixed time budget (for example, 1-2 hours). If the task does not fit:

- Shrink the scope, or
- Split it into smaller refactoring tasks.

This prevents "infinite cleanup" from consuming research momentum.

### Why this matters

Bounded refactoring:

- Reduces onboarding time
- Lowers cognitive load
- Cuts debugging effort
- Improves reproducibility
- Makes future extensions easier

You may not be paid to refactor, but you always pay for not refactoring.

## Exercises

By now, you likely have a strong refactoring candidate:

- A Python script
- A module or notebook inherited from a colleague
- Any code you repeatedly struggle to read, debug, or extend

Instructions:

- Choose one target.
- Refactor incrementally without changing behaviour.
- Use version control and/or minimal tests to validate changes.

> **Tip:** Revisit the "Top tips" lecture and define a realistic time-box before you start.

Afterwards, reflect:

- What was your most impactful change?
- Which anti-patterns did you remove?
- Did future changes become easier or safer?
- Did you stay within your time-box? If not, why?
