# Writing new code

## Scenarios

Calling this lecture "Writing new code" is a little misleading. In research, we almost never start from a true blank slate: even brand-new projects rely on libraries, frameworks, and tools built by other people. That is not a weakness&mdash;it is exactly how modern software works. Reusing reliable foundations lets you focus on the science instead of rebuilding infrastructure.

For this lecture, assume you are starting something relatively fresh (a new project, prototype, or exploratory path). The same principles still apply when extending or reshaping legacy code, especially when you know that codebase well enough to make structural decisions.

In practice, you will usually be in one or more of these scenarios:

- **Pipeline/workflow development**: Combine existing tools into a reproducible process. Priority: clear structure and robustness over clever scripting.

- **Interface/glue code**: Connect independent codebases or tools. Priority: explicit contracts (inputs, outputs, assumptions) and defensive error handling.

- **High-performance scientific code**: Implement custom numerical methods when existing tools are not enough. Priority: correctness, numerical stability, and maintainability (plus performance where needed).

- **Reusable tool/library development**: Build code others will reuse (including future you). Priority: modular design, stable interfaces, testing, and documentation.

- **Exploratory/prototyping code**: Test ideas quickly. Priority: fast iteration now, with refactoring later if the prototype becomes important.

Across all scenarios, the aim is the same: write code that is clear, reliable, and easy to evolve.

## From science to code

Mapping a research question to code is a distinct skill from coding itself. As a domain expert, you are often best placed to do this translation: define goals, constraints, and acceptable behaviour before writing substantial code.

The earlier lectures on object-oriented, interface, and functional design provide your building blocks. Use them to sketch structure first, then implement.

1. **Tangible objects**: Identify real entities in the problem domain (for example, a planet with mass and radius) and map them to intuitive code objects.

2. **Abstract objects**: Define computational abstractions that are useful even if they are not physical entities (for example, a `Reaction` object in a reaction-network model).

3. **Operations on objects**: Decide what actions belong as methods versus standalone functions, based on cohesion and clarity.

4. **Interaction/dependency map**: Sketch how components depend on each other. Minimise unnecessary cross-talk and hidden coupling.

Then ask: *Will this structure support what I need to do?* If yes, build a minimum working example. If not, iterate on the design before writing more code.

## A common design tension

When writing new code, there is a recurring tension between two failure modes:

- A quick bespoke solution that only meets today's requirement.
- Over-engineering for hypothetical future needs.

In academia, this tension is amplified by time pressure. People often get trapped in one of two loops:

1. "I can't refactor because I need results."
2. "I can't add features because I'm still cleaning code."

Both slow progress. The goal is balance: meet current needs clearly while keeping the code adaptable for likely next steps, including handover to new team members.

## Guiding principles

To manage that tension, use a phased approach.

### Phase 1

**Make it work, and make it safe.**

1. **Build a minimum working example (MWE)**: Prove the core idea is scientifically useful before polishing everything.

1. **Anchor the MWE with tests**: Even minimal tests give you a behavioural baseline.

1. **Refactor early**: Apply DRY and basic structure improvements before shortcuts harden into debt.

The outcome of Phase 1 should be a small, useful, test-anchored codebase with a clear structure.

### Phase 2

**Make it easier to change.**

1. **Use the MWE directly when sufficient**: Sometimes the minimal version already supports real science.

1. **Extend deliberately**: As new goals emerge, add features incrementally.

If a change feels hard, treat that as a design signal. Refactor first (structure, responsibilities, interfaces), then implement the new behaviour.

The outcome of Phase 2 is active development with controlled complexity.

### Phase 3

**Make it stable and transferable.**

Typical priorities:

1. **Strengthen tests and validation**: Cover edge cases, failure modes, and representative scientific use cases.

1. **Clarify interfaces and assumptions**: Make expectations explicit through signatures, docstrings, and lightweight docs.

1. **Improve reproducibility and usability**: Pin dependencies, standardise configuration, improve examples and errors.

1. **Prepare for handover/publication**: Ensure knowledge lives in the codebase, not only in one person's memory.

The outcome of Phase 3 is a reliable codebase that survives personnel turnover.

### At a glance

- **Phase 1**: Make it work
- **Phase 2**: Make it easy to change
- **Phase 3**: Make it stable and shareable

## API and library layer

Modularisation and encapsulation usually show up as two layers:

> *Definition*: An **Application Programming Interface (API)** is the user-facing layer: the entry points users call.

> *Definition*: The **library layer** is the implementation layer: internal logic users typically should not interact with directly.

Start separating these early (often in Phase 1). A cleaner boundary makes refactoring safer and later scaling easier.

## Exercises

1. **Conceptual mapping**: Choose a fresh/minimal project (or a research problem in your domain) and map the problem to code structure before writing substantial code.

    > **Tip**: You can do this entirely conceptually. Focus on tangible/abstract objects, interactions, and responsibilities.

2. **Minimum working example (MWE)**: Implement a simple prototype and verify basic correctness with sanity checks or small tests.
