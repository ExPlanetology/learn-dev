# Object-oriented programming

I’ll start with a core OOP idea: **classes**. This is not a claim that OOP is always better than other paradigms. In science, classes are often avoided simply because they feel unfamiliar, not because they are a bad fit. So the aim here is practical: make classes understandable, useful, and less intimidating.

> *Definition*: **Object-oriented programming (OOP)** is a paradigm that organises software around objects&mdash;created from classes&mdash;that combine data (attributes) with behaviour (methods).

Classes can improve organisation, readability, and extensibility, especially as projects grow. More importantly, they give you a useful mental model: think of your problem as interacting components with clear responsibilities. Once that model is clear, syntax is the easy part.

We will use a top-down approach: create a class first, then unpack how it works.

## Class definition

Let's define a `Dog` class. Think of a class as a blueprint: it describes what a dog *has* (attributes) and what a dog *does* (methods).

```python
class Dog:
    """A dog

    Args:
        name: Name
        age: Age

    Attributes:
        species: Species
        name: Name
        age: Age
    """

    # Class attribute (shared across all Dog instances unless overridden)
    species: str = "Canis familiaris"

    def __init__(self, name: str, age: float) -> None:
        self.name: str = name  # Instance attribute
        self.age: float = age  # Instance attribute

    def speak(self) -> str:
        """Barks"""
        return f"{self.name} says woof!"

    def say_hello(self, friend: str) -> str:
        """Say hello to a friend"""
        return f"{self.name} the dog says: Woof! Hello {friend}!"
```

**Explanation**

1. **Class definition**

    - `class Dog:` defines a blueprint for creating dog objects.
    - `species` is a **class attribute** (shared by all instances unless overridden on one instance).

2. **Initialiser (`__init__`)**

    - `__init__` runs automatically when you create an instance.
    - `self` refers to the current instance.
    - `name` and `age` are inputs, stored as `self.name` and `self.age`.
    - `__init__` is a Python "dunder" (double-underscore) method.

3. **Instance methods**

    - `speak` and `say_hello` are methods that operate on one specific dog instance.
    - Because methods access `self`, output can depend on that dog's current state (for example, its name).

That was a lot of terminology for a first class, so if it feels dense, that is completely normal.

Now let’s step back and *use* the class in practice.

## Using classes

We'll begin by creating an instance of the `Dog` class with the name `"Rocket"` and age `3`:

```python
my_dog: Dog = Dog("Rocket", 3)
```

We can access the attributes of the instance:

```python
name: str = my_dog.name
age: float = my_dog.age
print("name: ", name)
print("age: ", age)
```

We can ask `my_dog` to `speak`:

```python
my_dog.speak()
```

And say hello to a friend:

```python
my_dog.say_hello("Marie Curie")
```

We can also modify attributes and call `speak()` again. The output updates immediately because methods read the object's current state at call time.

```python
my_dog.name = "Rakete"
my_dog.speak()
```

This illustrates **encapsulation**, one of the main strengths of classes. Encapsulation works at the object level; **modularity** works at the system level. Together they make code easier to reason about and safer to change.

> *Definition*: **Encapsulation** bundles data and methods into one unit (for example, a class), exposing what users need while hiding internal details.

> *Definition*: **Modularity** organises a programme into distinct, self-contained components.

Key benefits:

1. **Separation of concerns**: Clear responsibilities and cleaner boundaries.

2. **Reusability**: Components can be reused across files or projects.

3. **Maintainability**: Changes are easier when behaviour is localised.

4. **Testability**: Components can be tested in isolation.

In our example, `Dog` has a clear responsibility (dog-related data and behaviour), and we can reuse it to create many dog instances:

```python
barney: Dog = Dog("barney", 3)
cyril: Dog = Dog("cyril", 6)
trevor: Dog = Dog("trevor", 10)
```

`Dog` also improves maintainability and testability because the behaviour is self-contained. If we add a dog-related feature, we usually add it in one place: the `Dog` class.

## Class vs instance attributes

This is a common source of confusion in Python OOP.

When we defined `Dog`, we added a class attribute called `species`. Accessing `my_dog.species` gives the value from the class:

```python
my_dog.species
```

Now let's update the class attribute on `Dog` itself:

```python
Dog.species = "Alien dog"
```

This change is reflected in `my_dog`, which was an object created from `Dog`:

```python
my_dog.species
```

So where does confusion begin? It starts when we assign `species` on `my_dog` (an instance). At that point, we create an **instance attribute** that shadows the class attribute.

```python
my_dog.species = "Non felem"
my_dog.species
```

And if we look again at `Dog.species`, we can see that its attribute hasn't changed&mdash;which is expected, because we've only overridden the attribute on the `my_dog` instance, not on the class itself:

```python
Dog.species
```

This interaction between class and instance attributes touches two core OOP ideas: **inheritance** (instances receive behaviour/data from a class hierarchy) and **overriding/shadowing** (instance-level values can replace class-level ones). These features are powerful, but they also cause many beginner bugs when used unintentionally.

> **Note:** In Python, you can create attributes dynamically (for example, `my_dog.new_attribute = value`). This flexibility is useful in limited cases, but it is often discouraged in structured projects because it bypasses the class design and can make code harder to reason about.

## Defining your own classes

Hopefully this makes classes feel less mysterious. A common next question is: *How many classes should I create?* That question is central to software design.

Be careful of creating too many classes, which can lead to **class explosion**.

> *Definition*: A **class explosion** is uncontrolled growth in the number of classes, usually caused by over-engineering. The result is a codebase that is harder to read, navigate, and maintain.

Start simple. Classes can represent data, behaviour, or both; they can also compose other classes. Begin with the few classes that give immediate structural benefit, then expand gradually.

Also remember: classes do not need to represent only tangible things (like dogs or animals). As designs mature, many useful classes represent abstract concepts and interactions.
