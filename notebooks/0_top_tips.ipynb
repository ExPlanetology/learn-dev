# Top tips

**If you're short on time, make this the part you actually read and take in&mdash;it'll pay off**. Getting the basics right in software development doesn't just make your own life easier&mdash;it makes your code a joy for others to work with. Whether you're collaborating with colleagues or calling in professional software engineers, clean, well-structured code means they can focus on the hard problems rather than untangling messy logic or tracking down bugs born from poor design choices.

## Foundational tips

These are the essentials you should adopt immediately to have a strong foundation:

1. **Version control**: In practice, it's nearly impossible to work on any software project without using a version control system, and Git is by far the most popular. Git lets you track every change, maintaining a complete history of your project's development. It works across all major platforms and is highly recommended for managing code efficiently and safely.

1. **Modularity**: Break your code into clearly defined modules, each responsible for a specific task or concern. This is the principle of *separation of concerns* and makes your code easier to read, test, and maintain.

1. **Keep functions and methods small**: Each function or method should do one thing and do it well. Smaller, focused functions are easier to test, debug, and reuse.

1. **Consistent naming conventions**: Use clear, descriptive names for variables, functions, and classes. Stick to a consistent style throughout your project to improve readability. For example, use `snake_case` or `camelCase` consistently, and avoid vague names like `data` or `temp` unless absolutely necessary.

1. **Avoid global variables**: Avoid relying on global variables. They make code harder to understand, debug, and maintain because any part of your program can change them, sometimes in unexpected ways. Instead, pass variables explicitly to the functions or classes that need them. The less hidden cross-talk between different parts of your program, the fewer mysterious bugs you'll have to hunt down.

1. **Use optimised libraries**: There is an entire ecosystem of libraries that already implement standard algorithms, often highly optimised, well-tested, and far more efficient than a quick in-house version. Use them, and design your code to take advantage of them from the outset. For Python-based work this means NumPy, SciPy, JAX, and their equivalents: prefer vectorised operations, avoid unnecessary array copies, and let the library do the heavy lifting. Reinventing the wheel wastes time and introduces subtle bugs.

1. **Tests**: Write small test snippets to serve as an *anchor*: something you know works and must not break. Build tests as you develop to catch problems early.

1. **Documentation and comments**: At a minimum, comment your code to explain *why* something is done, while the code itself shows *how*. Consider adding *docstrings* for functions, classes, and modules to provide structured, standardised documentation.

## Intermediate tips

These practices improve maintainability, collaboration, and code quality:

1. **Measure twice, cut once**: Before writing any code, invest time in understanding *exactly what you want to accomplish*. This step is not always easy, but it is precisely where humans have a key advantage over AI: we can reason about context, nuance, and trade-offs in ways that AI tools cannot. Clarifying the goal, constraints, and desired behaviour prevents wasted effort and unnecessary complexity. In practice, you may need to revisit and refine your understanding as you go, but every iteration will save far more time than it costs.

1. **Avoid deep nesting and complex logic**: Keep control flow simple and flat. Deeply nested loops or conditionals make code hard to read and understand. Consider early returns or breaking complex logic into smaller helper functions.

1. **Use standard formats for inputs and outputs**: When handling data, prefer widely supported, standardised formats instead of inventing custom ones. For inputs, formats like JSON, YAML, or XML make your configuration files human-readable and machine-parseable. For outputs, formats such as CSV, NetCDF, or HDF5 provide efficient storage, compression, and interoperability with existing tools. Sticking to standards ensures long-term readability, easier sharing with collaborators, and compatibility with established analysis and visualisation software.

1. **Incremental refactoring**: Refactoring is the process of improving code structure&mdash;removing duplication, clarifying responsibilities, simplifying interfaces&mdash;without changing external behaviour. When a modification feels difficult, that is usually a signal that the surrounding code is not structured well for change. Rather than forcing the change, first refactor: clean up the structure and simplify the design. The change you originally wanted will often become trivial afterwards. When working incrementally on a larger codebase, take every small opportunity to clean up along the way. These improvements accumulate steadily, following the **Don't Repeat Yourself (DRY)** principle and leaving the code healthier than you found it.

1. **Handle errors gracefully**: Don't ignore potential errors or edge cases. Use proper error handling, like exceptions or error codes, to ensure your program can recover or at least fail informatively. Additionally, use assert statements to check validity assumptions during development, and complement this with logging to record important events or warnings at runtime.

1. **Use code reviews**: Whenever possible, get a second pair of eyes on your code. Code reviews catch bugs, improve style, and spread knowledge across the team, leading to higher-quality and more maintainable code.

## Advanced tips

These are strategies for long-term maintainability, professional practices, and optimisation:

1. **Automate repetitive tasks**: Use scripts, build tools, or continuous integration (CI) pipelines to automate testing, deployment, formatting, or other routine tasks. This saves time, reduces human error, and keeps your workflow consistent.

1. **Avoid premature optimisation**: Write clear, correct code first. Optimise only when you have evidence from profiling data that performance is actually a bottleneck. Early micro-optimisations make code harder to read and maintain without meaningful benefit. The key question to ask is: *"Is the speed of my code actually preventing me from doing research?"* More often than not, it is not the real bottleneck.

1. **Code rots**: Code is never truly "finished". Programming languages, libraries, and best practices evolve continuously, and dependencies on external code can change or introduce security risks. Static code gradually becomes outdated and harder to maintain meaning that inaction is effectively moving backwards. If you want your code to remain a valuable long-term resource, budget time regularly for maintenance and updates.

## Enjoy the ride

Writing clean, maintainable code isn't just a technical skill&mdash;it's a mindset and an investment. The time you spend upfront organising, documenting, and testing your code pays off many times over in reduced headaches, faster development, and more confident collaboration down the line. **Remember, code is a form of communication&mdash;not just to machines, but to your future self and others who will read, use, and build upon your work.** Write your code as if the person maintaining it is a ruthless and impatient stranger who must understand it quickly (because often, that person *is* you in a few months).

Even with the rise of AI-assisted coding tools, it remains essential that you deeply understand what your code is trying to achieve. **AI can help write or optimise code, but it cannot replace your insight into the problem domain, the intended logic, or the design decisions behind your implementation.** Without this understanding, it is easy to produce code that works superficially but fails when faced with edge cases, evolving requirements, or integration challenges. Embrace continuous improvement, don't be discouraged by technical debt or legacy code, and view every messy codebase as an opportunity to leave things better than you found them.

Finally, enjoy the process. **Software development is creative problem-solving at its core.** The better your foundation, the more time and energy you'll have for innovation, experimentation, and solving the truly challenging problems.
