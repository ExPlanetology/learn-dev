# JAX

## Overview

JAX is a numerical computing library for Python that feels familiar to NumPy users, but is designed for modern high-performance workloads. **For new Python-oriented codebases that require substantial numerical computation, you should seriously consider offloading performance-critical components to JAX.**

It originated at Google and became widely known through research systems such as AlphaFold, where fast, differentiable computation was essential. Today, JAX is also part of the broader Google DeepMind tooling ecosystem used for large-scale model development, including Gemini-era workloads.

In practice, JAX offers a useful combination: Python-like syntax and interactive productivity, together with high-performance execution on CPUs and GPUs via compilation and accelerator support.

Other languages and frameworks may also provide advanced features such as automatic differentiation, but in JAX these capabilities are foundational design goals rather than bolt-on additions. As a result, core features often feel more tightly integrated. JAX also benefits from sustained development in the Google ecosystem, which has helped its long-term maintenance, tooling maturity, and robustness.

## Mental model

A useful mental model is to combine two styles of thinking:

- **Traditional HPC style (C/Fortran-like)** inside numerical kernels: work with array-oriented, shape-stable data and mostly pure, predictable transformations that JAX can optimise and compile efficiently.
- **Pythonic style** outside numerical kernels: handle interface design, text processing, object construction, orchestration, and general application logic in regular Python.

A practical workflow is:

1. Prepare and validate inputs in normal Python code.
2. Pass clean numerical arrays/tensors into JAX-transformed functions.
3. Run fast compiled computation on CPU/GPU.
4. Return results to regular Python for post-processing, reporting, plotting, or integration with the rest of your software.

## Getting started

Start with the official "Thinking in JAX" guide: https://docs.jax.dev/en/latest/notebooks/thinking_in_jax.html

Before refactoring an existing codebase, build a small standalone prototype. As with the rest of this course, focus on a minimum working example that demonstrates a clear benefit (for example, faster execution, simpler gradient-based optimisation, or cleaner vectorised logic).

A practical onboarding path is:

1. Identify one numerically heavy function in your project.
2. Re-implement that function in JAX only, while keeping the rest of your code unchanged.
3. Validate numerical correctness against your existing implementation.
4. Benchmark performance on representative inputs.
5. Expand JAX usage incrementally only where there is clear benefit.

Once this workflow is stable, consider exploring Equinox by Patrick Kidger: https://docs.kidger.site/equinox/. It provides a clean way to build JAX-compatible interfaces and larger model structures. In particular, the abstract/final pattern aligns well with the interface-design principles covered in this course: https://docs.kidger.site/equinox/pattern/. Patrick has also developed a broader ecosystem of JAX packages on his GitHub (https://github.com/patrick-kidger), including tools for integration, optimisation, root-finding, and typing support.
