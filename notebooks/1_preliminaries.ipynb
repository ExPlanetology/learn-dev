# Preliminaries

Before we begin the course, there are a few preliminary topics I'd like to briefly cover to set the stage. Further reading is provided for those interested in a deeper discussion. The aim here is to introduce the material as concisely as possible, so you can quickly move on to the main content.

## Interactive development environment (IDE)

Before diving into any coding or software project, it's important to use an interactive development environment (IDE). IDEs provide a range of powerful tools that can significantly improve your productivity and workflow. There are many options available, and over time you’ll develop your own preferences for how to configure your environment. You can find an overview of the setup I recommend in `additional_reading/IDE_setup.ipynb`.

## Git

Git is a fundamental tool for modern software development. It enables you to track changes to your code over time, collaborate with others without overwriting each other's work, and maintain a history of your project’s evolution. Whether you're working alone or as part of a team, version control is essential for managing code in a structured and reproducible way.

While Git is a powerful and widely used system, this course will not go into depth on how to use it. Fortunately, there are many excellent tutorials and guides available online to help you get started, for example:

- https://product.hubspot.com/blog/git-and-github-tutorial-for-beginners
- https://rogerdudler.github.io/git-guide/
- https://docs.github.com/en

## Documentation

Documentation is often one of the last considerations in academic or research code. However, in this course, we will make it a priority by embedding documentation directly into class definitions using **docstrings**. Similarly, methods will include their own docstrings to explain their purpose and usage. Occasionally, we’ll omit docstrings in examples to keep code snippets compact and focussed, but the habit of documenting your code should be consistent in your own work.

*Further reading:* `additional_reading/documentation.ipynb`

## AI assistants

Using AI tools with simple prompts is an excellent way to generate template code that you can modify for your specific needs. It’s also a valuable resource for debugging, explaining unfamiliar code, and accelerating your learning process. Think of it as a powerful assistant for rapid iteration and exploration. For this reason, I don't provide unnecessary or specific code examples in this course, since AI can generate these for you. Instead, the **focus is on core concepts, so that you can understand, critique, and adapt AI-generated code** to suit your particular goals.
